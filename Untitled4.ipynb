```python
!pip install datasets pandas
from datasets import load_dataset

# Загружаем данные
dataset = load_dataset("ahmetozcan/codedocia25")

# Превращаем в таблицу для удобства
df = dataset['train'].to_pandas()

# Выводим первые 5 строк, чтобы увидеть данные
print(df.head())
```

```python
X = df.drop(columns=['Satış Cirosu', 'Gün'])
```

```python
import pandas as pd  # Добавляем это в начало

# 1. Превращаем всё в числа, а то, что не число (как наш '?') — станет "пустотой"
X = X.apply(pd.to_numeric, errors='coerce')

# 2. Заполняем все пустоты нулями
X = X.fillna(0)
```

```python
y = df['Satış Cirosu']
```

```python
# Оптимальные параметры для 1.1 млн строк на бесплатном Colab:
# n_estimators=40 — достаточно для стабильности
# max_depth=15 — позволит модели выучить сложные зависимости (праздники, курсы)
# min_samples_leaf=10 — защитит от переобучения и сэкономит память
model = RandomForestRegressor(
    n_estimators=40,
    max_depth=15,
    min_samples_leaf=10,
    random_state=42,
    n_jobs=-1
)

print("Запускаю оптимизированное обучение...")
model.fit(X, y)
print("--- ПОБЕДА! МОДЕЛЬ ОБУЧЕНА И СТАБИЛЬНА ---")
```

```python
# Просим ИИ сделать предсказание на основе наших данных X
predictions = model.predict(X.head())

# Выводим предсказания ИИ
print("Предсказания ИИ:", predictions)

# Выводим реальные цифры из таблицы для сравнения
print("Реальные продажи:", y.head().values)
```

```python
import matplotlib.pyplot as plt

# 1. Считаем важность каждого столбца
importances = model.feature_importances_
feature_names = X.columns

# 2. Создаем красивый график
plt.figure(figsize=(10, 6))
plt.barh(feature_names, importances, color='skyblue')
plt.xlabel('Важность для предсказания')
plt.title('На что больше всего смотрит наш ИИ?')
plt.show()
```

```python
# 1. Берем одну случайную строку из данных (имитируем новые данные)
sample_data = X.sample(1)

# 2. Просим ИИ предсказать продажи для этой строки
tahmin = model.predict(sample_data)

# 3. Выводим результат красиво
print("--- РЕЗУЛЬТАТ МОДЕЛИ (Satış Tahmini) ---")
print(f"Прогноз продаж для этого магазина: {tahmin[0]:.2f} лир")
print("-" * 40)

# Показываем, на основе каких данных сделан прогноз
print("Данные для этого прогноза:")
print(sample_data)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor

# --- 1. FEATURE ENGINEERING (Создаем интеллект модели) ---
df['Gün'] = pd.to_datetime(df['Gün'])
df['Month'] = df['Gün'].dt.month
df['DayOfWeek'] = df['Gün'].dt.dayofweek
df['is_weekend'] = df['DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)

# Погода (средняя температура по месяцам в TR)
temp_map = {1:5, 2:7, 3:10, 4:15, 5:20, 6:25, 7:28, 8:28, 9:23, 10:18, 11:12, 12:8}
df['avg_temp'] = df['Month'].map(temp_map)

# Экономика (Курс USD/TRY 2021)
usd_rates = {1:7.4, 2:7.1, 3:7.6, 4:8.2, 5:8.3, 6:8.6, 7:8.6, 8:8.5, 9:8.5, 10:9.1, 11:10.5, 12:13.5}
df['usd_rate'] = df['Month'].map(usd_rates)

# Праздники 2021
holidays = ['2021-01-01', '2021-04-23', '2021-05-01', '2021-05-12', '2021-05-13',
            '2021-05-14', '2021-05-15', '2021-05-19', '2021-07-15', '2021-07-19',
            '2021-07-20', '2021-07-21', '2021-07-22', '2021-07-23', '2021-08-30', '2021-10-29']
df['is_holiday'] = df['Gün'].isin(pd.to_datetime(holidays)).astype(int)

# --- 2. ОЧИСТКА И ПОДГОТОВКА (Data Cleaning) ---
y = df['Satış Cirosu']
X = df.drop(columns=['Satış Cirosu', 'Gün'])
X = X.apply(pd.to_numeric, errors='coerce').fillna(0)

# --- 3. ОБУЧЕНИЕ МОЩНОЙ МОДЕЛИ ---
# n_estimators=30 и max_depth=10 — это идеальный баланс для памяти твоего устройства
model = RandomForestRegressor(n_estimators=30, max_depth=10, random_state=42, n_jobs=-1)

print("Начинаю глубокое обучение модели...")
model.fit(X, y)
print("--- ГОТОВО! МОДЕЛЬ ПОЛНОСТЬЮ ОБОРУДОВАНА ---")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- ШАГ 1: Настройка визуализации ---
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

# --- ШАГ 2: Подготовка данных о важности факторов ---
importances = model.feature_importances_
feature_names = X.columns

# Создаем таблицу данных для графика
fi_df = pd.DataFrame({'Faktör': feature_names, 'Önem Düzeyi': importances})
fi_df = fi_df.sort_values(by='Önem Düzeyi', ascending=False)

# --- ШАГ 3: Построение графика на турецком ---
# Используем палитру 'viridis' для профессионального вида
sns.barplot(x='Önem Düzeyi', y='Faktör', data=fi_df, palette='viridis')

# Заголовки на турецком для жюри
plt.title('Satış Tahmininde Etкиleyen Faktörlerin Analizi', fontsize=16) # Анализ факторов, влияющих на прогноз продаж
plt.xlabel('Önem Derecesi', fontsize=12) # Степень важности
plt.ylabel('Veri Özellikleri', fontsize=12) # Признаки данных

plt.show()

# --- ШАГ 4: Финальный текстовый вывод на турецком ---
print("\n" + "="*50)
print("MODEL DEĞERLENDİRME SONUÇLARI") # Результаты оценки модели
print("="*50)
print(f"Eğitilen veri satırı sayısı: {len(df)}") # Количество обученных строк
print("Eklenen Özellikler: Enflasyon (USD), Bayramlar, Hava Durumu") # Добавленные фичи: Инфляция, Праздники, Погода
print("Durum: Tahminleme için hazır!") # Статус: Готов к прогнозированию!
print("="*50)
```

```python
from sklearn.metrics import r2_score

# --- 1. Просим ИИ сделать предсказание на тех данных, что у нас есть ---
predictions = model.predict(X)

# --- 2. Считаем коэффициент R2 (тот самый результат от 0 до 1) ---
score = r2_score(y, predictions)

# --- 3. Выводим на турецком для жюри ---
print("\n" + "="*50)
print("MODEL PERFORMANS ANALİZİ (R² Skoru)") # Анализ производительности модели
print("="*50)
print(f"Başarı Oranı: {score:.4f}") # Коэффициент успеха
print(f"Yüzdelik Doğruluk: %{score*100:.2f}") # Точность в процентах
print("="*50)

if score >= 0.85:
    print("Sonuç: Mükemmel! (Отличный результат)")
elif score >= 0.70:
    print("Sonuç: İyi, geliştirilebilir. (Хорошо, можно улучшить)")
else:
    print("Sonuç: Daha fazla veri veya özellik gerekiyor. (Нужно больше данных)")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# --- 1. MAE (Ortalama Mutlak Hata) hesaplama ---
mae = mean_absolute_error(y, predictions)

# --- 2. RMSE (Ortalama Karesel Hatanın Karekökü) hesaplama ---
rmse = np.sqrt(mean_squared_error(y, predictions))

# --- 3. Sonuçları Türkçe olarak yazdırma ---
print("\n" + "="*50)
print("MODEL HATA METRİKLERİ")
print("="*50)
print(f"Ortalama Mutlak Hata (MAE): {mae:.4f}")
print(f"Ortalama Karesel Hatanın Karekökü (RMSE): {rmse:.4f}")
print("="*50)

# Küçük bir yorum ekleyelim
if mae < y.mean() * 0.1 and rmse < y.mean() * 0.15: # Örnek eşikler, gerçek eşikler veri setine göre değişir
    print("Yorum: Hata değerleri oldukça düşük, model performansı başarılı görünüyor!")
else:
    print("Yorum: Hata değerleri incelenmeli, potansiyel iyileştirme alanları olabilir.")
print("="*50)
```

```python
# --- 1. ПРОВЕРКА И ВОССТАНОВЛЕНИЕ ДАННЫХ ---
# Если мы раньше удалили Mağaza Kodu, этот код поможет его не потерять
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error

# Сортируем данные для расчетов (важно для "памяти")
# Если Mağaza Kodu нет, код просто пропустит сортировку по нему
try:
    df = df.sort_values(by=['Mağaza Kodu', 'Gün'])
    group_col = 'Mağaza Kodu'
except:
    df = df.sort_values(by=['Gün'])
    group_col = None

# --- 2. СОЗДАЕМ МОЩНУЮ "ПАМЯТЬ" (Lag Features) ---
# Продажи вчера
df['yesterday_sales'] = df['Satış Cirosu'].shift(1)
# Продажи неделю назад (ритм выходных)
df['last_week_sales'] = df['Satış Cirosu'].shift(7)
# Среднее за неделю (плавный тренд)
df['weekly_trend'] = df['Satış Cirosu'].rolling(window=7).mean()

# Заполняем пустоты средними значениями
df = df.fillna(df['Satış Cirosu'].mean())

# --- 3. ПОДГОТОВКА ДЛЯ ИДЕАЛЬНОЙ ТОЧНОСТИ ---
y = df['Satış Cirosu']
# Оставляем только те столбцы, которые ИИ может превратить в числа
X = df.select_dtypes(include=[np.number]).drop(columns=['Satış Cirosu'], errors='ignore')

# --- 4. ОБУЧЕНИЕ (Усиленные параметры) ---
# Увеличим кол-во деревьев до 60 (Colab должен вытянуть с новой логикой)
model = RandomForestRegressor(n_estimators=60, max_depth=20, min_samples_leaf=3, random_state=42, n_jobs=-1)

print("Начинаю финальный тюнинг для минимизации ошибки...")
model.fit(X, y)

# --- 5. РЕЗУЛЬТАТЫ НА ТУРЕЦКОМ ---
preds = model.predict(X)
final_r2 = r2_score(y, preds)
final_mae = mean_absolute_error(y, preds)

print("\n" + "="*50)
print("HEDEFİMİZ: MİNUMUM HATA (ЦЕЛЬ: МИНИМУМ ОШИБОК)")
print("="*50)
print(f"Yeni Başarı Oranı (R²): %{final_r2*100:.2f}")
print(f"Yeni Ortalama Hata (MAE): {final_mae:.2f} TL")
print(f"Hata payı %{(final_mae/df['Satış Cirosu'].mean())*100:.2f} oranına düşürüldü!")
print("="*50)
```
